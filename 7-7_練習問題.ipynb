```python
# (2)円周率3.141592について、小数点以下第1位から小数点第5位を四捨五入した値をそれぞれ入力する
pi = 3.141592
c = 0
for i in range(5):
    print(f'小数点第{c}位を四捨五入した値は{round(pi,c-1)}です')
    c += 1
```

```python
# (2)円周率3.141592について、小数点以下第1位から小数点第5位を四捨五入した値をそれぞれ入力する
pi = 3.141592
c = 0
for i in range(5):
    print(f'小数点第{c+1}位を四捨五入した値は{round(pi,c)}です')
    c += 1
```
