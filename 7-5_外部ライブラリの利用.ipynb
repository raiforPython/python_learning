```python
# requests でPythonの公式サイトを取得する
import requests

response = requests.get('https://www.python.org/downloads/')
text = response.text
print(text)
```

```python
# 標準ライブラリを用いたwebページの取得
import http.client

conn = http.client.HTTPConnection('www.python.go')
conn.request('get','/downloads/')
response = conn.getresponse()
text = response.read().decode('UTF-8')
print(text)
conn.close
```

```python
# 標準ライブラリを用いたwebページの取得
import http.client

conn = http.client.HTTPConnection('www.python.org')
conn.request('GET','/downloads/')
response = conn.getresponse()
text = response.read().decode('UTF-8')
print(text)
conn.close
```
